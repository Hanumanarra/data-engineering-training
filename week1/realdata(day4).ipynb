```python
%sql
 Use ROW_NUMBER() OVER (PARTITION BY student_id ORDER BY timestamp) to identify duplicates.
```

```python
%sql
select 
    student_id,
    email_address,
    timestamp,
    case
        when rn=1 then 'Valid Submissions'
        else 'Duplicates'
    
    end as submission_status
from(
    select 
    m.student_id,
    r.Email_address,
    r.timestamp,
    row_number() over(partition by student_id order by timestamp) as rn
    from df2 r
    Inner join df0_mail_map  m 
    on lower(r.Email_address)=lower(m.email)
) t
order by student_id,timestamp
```

```python
KEEP FIRST SUBMISSION,MARK OTHER AS DUPLICATES
```

```python
%sql
select
  m.student_id,
  r.email_address,
  r.timestamp,
  
  case
      when row_number() over(partition by m.student_id order by r.timestamp)=1
      then 'Submitted'
      else 'Duplicate'
  end as submission_status
  from df2 r
  inner join df0_mail_map m
  on lower(r.Email_address)=lower(m.email)
  order by m.student_id,r.timestamp;
```

```python
%sql
SELECT
       m.student_id,
       COUNT(DISTINCT LOWER(r.email_address)) AS email_used_count

FROM df2 r

INNER JOIN df0_mail_map m
ON LOWER(r.Email_address) = LOWER(m.email)

GROUP BY m.student_id

HAVING COUNT(DISTINCT LOWER(r.email_address)) > 1;
```

```python
Generate final classification: Submitted, Not Submitted, Duplicate, Invalid.
```

```python
WITH valid_submissions AS (
        SELECT
           m.student_id,
           r.email_address,
           r.timestamp,

           ROW_NUMBER() OVER (
               PARTITION BY m.student_id
               ORDER BY r.timestamp
           ) AS rn

    FROM df2 r

    JOIN df0_mail m
    ON LOWER(r.email_address) = LOWER(m.email)

),

classified_valid AS (

    SELECT
           student_id,
           email_address,
           timestamp,

           CASE
                WHEN rn = 1 THEN 'Submitted'
                ELSE 'Duplicate'
           END AS final_status

    FROM valid_submissions
),

invalid_submissions AS (

    SELECT
           NULL AS student_id,
           r.email_address,
           r.timestamp,
           'Invalid' AS final_status

    FROM response_2 r

    LEFT JOIN mail_map_table m
    ON LOWER(r.email_address) = LOWER(m.email)

    WHERE m.student_id IS NULL
),

not_submitted AS (

    SELECT
           c.regno AS student_id,
           NULL AS email_address,
           NULL AS timestamp,
           'Not Submitted' AS final_status

    FROM cleaned_data_set c

    WHERE NOT EXISTS (

        SELECT 1
        FROM mail_map_table m
        JOIN response_2 r
        ON LOWER(m.email) = LOWER(r.email_address)

        WHERE m.student_id = c.regno
    )
)

SELECT * FROM classified_valid

UNION ALL

SELECT * FROM invalid_submissions

UNION ALL

SELECT * FROM not_submitted;
```
